```python
import timesynth as ts
```

```python
# Initializing TimeSampler
time_sampler = ts.TimeSampler(stop_time=20)
print(time_sampler)
```

```python
# Sampling irregular time samples
irregular_time_samples = time_sampler.sample_irregular_time(num_points=5, keep_percentage=100)
```

```python
# Sampling regular time samples
regular_time_samples = time_sampler.sample_regular_time(num_points=101)
```

```python
print(regular_time_samples)
```

```python
# Initializing Sinusoidal signal
sinusoid = ts.signals.Sinusoidal(amplitude=10,frequency=0.25)
```

```python
sinusoid2 = ts.signals.Sinusoidal(amplitude=10,frequency=0.50)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Plotting the series
plt.plot(regular_time_samples, sinusoid.sample_vectorized(regular_time_samples), marker='o', markersize=4)
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.title('Regularly sampled sinusoid without noise');
```

```python
# Plotting the series
combined_waves = sinusoid.sample_vectorized(regular_time_samples) + sinusoid2.sample_vectorized(regular_time_samples)
plt.plot(regular_time_samples, combined_waves, marker='o', markersize=1)
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.title('Regularly sampled sinusoid without noise');
```
